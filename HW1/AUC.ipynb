```python
import numpy as np
import matplotlib.pyplot as plt

rs = np.random.RandomState(4321)
p = 2
n = 200
py1 = 0.6
mean1 = np.r_[1,1.]
mean0 = -mean1

# These are the parameters learned through maximization from before.
w = np.r_[2.45641058, 1.55227045]
b = -0.824723538369
```

```python
# Generate some testing data
Y = (rs.rand(n) > py1).astype(int)
X = np.zeros((n,p))
X[Y==0] = rs.multivariate_normal(mean0, np.eye(p), size=(Y==0).sum())
X[Y==1] = rs.multivariate_normal(mean1, np.eye(p), size=(Y==1).sum())

# This is the model's prediction on the test data.
T = 1 / (1. + np.exp(-b-np.dot(X,w)))
```

```python
thresholds = [float(x)/100.0 for x in range(101)]
thresholds.reverse()
ROC =[[m* 0 for m in range(101)], [m*0 for m in range(101)]]

for j in range(101):
    
    t = thresholds[j]
    predicted = [1.0 if prob > t else 0.0 for prob in T]
    labels = [1.0 if y == 1 else 0.0 for y in Y]
    
    #print predicted
    #print labels
    
    get_count = lambda x: sum([1 for i in x if i is True])

    TP_t = get_count([predicted[i] == x and x == 1.0 for i, x in enumerate(labels)])
    TN_t = get_count([predicted[i] == x and x == 0.0 for i, x in enumerate(labels)])
    FP_t = get_count([predicted[i] == 1.0 and x == 0.0 for i, x in enumerate(labels)])
    FN_t = get_count([predicted[i] == 0.0 and x == 1.0 for i, x in enumerate(labels)])
    
    #if TP_t < 0.0000001:
    #    TP_t = 0.0
    #    TN_t = 0.0
    #    FP_t = 0.0
    #    FN_t = 0.0
    
    #print TP_t
    #print TN_t
    #print FP_t
    #print FN_t
    
    #print "________________"
    
    print "j is:", j
    
    # Compute false positive rate for current threshold.
    FPR_t = FP_t / float(FP_t + TN_t)
    ROC[0][j] = FPR_t

    # Compute true  positive rate for current threshold.
    TPR_t = TP_t / float(TP_t + FN_t)
    ROC[1][j] = TPR_t
    
AUC = 0.0
for i in range(100):
    AUC += (ROC[0][i+1] - ROC[0][i]) * (ROC[1][i+1] + ROC[1][i])
AUC *= 0.5

print AUC
    
    
    
    
    
    
```
